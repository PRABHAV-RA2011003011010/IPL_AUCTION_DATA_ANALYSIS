```python
import pandas as pd
import glob
import os

# Change year in path
directory_path = r'C:\Users\prabh\OneDrive\Desktop\IPL_Auction\All_Teams_Yearly_Data\2024_teams_data'

csv_files = glob.glob(os.path.join(directory_path, '*.csv'))
dataframes = []

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.head(50)
```

```python
#change year
combined_df['Year'] = 2024
print(combined_df.head(50))
```

```python
combined_df.drop('NATIONALITY', axis=1,inplace=True)
```
