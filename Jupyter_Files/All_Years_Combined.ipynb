```python
import pandas as pd
import glob
import os

#change path
directory_path = r'C:\Users\prabh\OneDrive\Desktop\IPL_Auction\combined_data_per_year'

csv_files = glob.glob(os.path.join(directory_path, '*.csv'))
dataframes = []

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dataframes.append(df)

final_df = pd.concat(dataframes, ignore_index=True)
print(final_df)
```

```python
my_dict = {
    0: 'Chennai Super Kings',
    1: 'Delhi Daredevils',
    2: 'Kings XI Punjab',
    3: 'Kolkata Knight Riders',
    4: 'Mumbai Indians',
    5: 'Rajasthan Royals',
    6: 'Royal Challengers Bangalore',
    7: 'Sunrisers Hyderabad'
}
```

```python
my_dict1617 = {
    'Chennai Super Kings': 'Delhi Daredevils',
    'Delhi Daredevils': 'Gujarat Lions',
    'Rajasthan Royals': 'Rising Pune Supergiant'
    
}
```

```python
final_df['Team'] = final_df['Team'].apply(lambda x: my_dict.get(x, x))
final_df
```

```python
final_df['Team'] = final_df.apply(
    lambda row: my_dict1617.get(row['Team'], row['Team']) if row['Year'] == 2016 else
                my_dict1617.get(row['Team'], row['Team']) if row['Year'] == 2017 else
                row['Team'],
    axis=1
)
final_df
```
